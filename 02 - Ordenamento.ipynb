# 6. Função final - Combinação de Todas Regras <a class="anchor" id="chapter6"></a>

A função abaixo satisfaz todas as condições abaixo, e é processada em menos de 5 segundos.

* Condição 1: O "domain_id" não pode se repetir em 4 posições consecutivas
* Condição 2: O "vertical" não pode se repetir em 1 posição consecutiva
* Condição 3: Se existir o id 641416750 ele deve ser colocado na posição 3, sendo esta regra mais forte que as demais
* Condição 4: Se existir o id 22351223 ele deve ser colocado na posição 6, sendo esta regra mais forte que as demais
* Condição 5: As posições 9,10,11 devem ter somente itens da categoria "HOME&DECO" sendo esta regra mais forte que as 1 e 2;
* Condição 6: Cumprindo todas estas condições, o ordenamento deve respeitar um ordenamento do maior para o menor escore.

```python
%%time
df= pd.read_csv("ordenamiento.csv")
df = df.sort_values('score', ascending = False)
df4 = df
lines_hd = [8, 9, 10]
id_p3 = 641416750
id_p6 = 22351223

df_p3 = df[df['item_id']==id_p3]
df_p6 = df[df['item_id']==id_p6]

if df_p3.shape[0] != 0:
    df4 = df4.loc[(df4["item_id"] != id_p3)]
if df_p6.shape[0] != 0:
    df4 = df4.loc[(df4["item_id"] != id_p6)]
    
for i in range(len(df4['vertical'])):
    vertical = df4['vertical'].values.flatten().tolist()
    domain   = df4['domain'].values.flatten().tolist()
    if i in lines_hd:
        w = i
        while domain[w] != 'MLC-HOME_DECOR':
            w = w+1
        temp = df4.iloc[w]
        df4.iloc[w] = df4.iloc[i]
        df4.iloc[i] = temp
        df4 = pd.concat([df4[0:i+1],df4[i+1:].sort_values('score', ascending = False)])
    elif vertical[i] == vertical[i-1]:
        k = i+1
        while vertical[i] == vertical[k]:
            k = k+1           
        temp = df4.iloc[k]
        df4.iloc[k] = df4.iloc[i]
        df4.iloc[i] = temp
        df4 = pd.concat([df4[0:i+1],df4[i+1:].sort_values('score', ascending = False)])
if df_p3.shape[0] != 0:
    df4 = pd.concat([df4[0:2], df_p3, df4[2:]])
if df_p6.shape[0] != 0:
    df4 = pd.concat([df4[0:5], df_p6, df4[5:]])
# df4.shape
df4.head(100)
```
